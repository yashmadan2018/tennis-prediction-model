# Tennis Prediction Model — Exploration

Use this notebook to explore the processed match data and prototype features.

```python
import sys
sys.path.insert(0, "..")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from utils.data_loader import load_processed

df = load_processed()
print(f"{len(df):,} matches loaded")
df.head()
```

```python
# Surface distribution
df["surface"].value_counts().plot(kind="bar")
plt.title("Matches by surface")
plt.tight_layout()
```

```python
# Matches per year
df["year"] = df["date"].dt.year
df.groupby(["year","tour"]).size().unstack().plot(kind="bar", stacked=True)
plt.title("Matches per year by tour")
plt.tight_layout()
```
